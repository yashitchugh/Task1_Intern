```python
#Handle missing values using mean/median/imputation

# Fill missing values in column 'Age' with its mean
df['Age'] = df['Age'].fillna(df['Age'].mean())

# Fill column 'Pclass' with its mean
df['Pclass'] = df['Pclass'].fillna(df['Pclass'].mean())
df.tail(10)
```

```python
#Convert categorical features into numerical using encoding.

from sklearn.preprocessing import LabelEncoder
# Label Encoding
le = LabelEncoder()
df['Cabin_encoded'] = le.fit_transform(df['Cabin'])
print(df)
```

```python
#Standarize numerical features (Z- Score Scaling)

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
# FIX: convert Series to DataFrame
pf = df[['Cabin_encoded']]  # double brackets make it 2D
df_scaled = pd.DataFrame(scaler.fit_transform(pf), columns=pf.columns)
print(df_scaled)
```

```python
#Visualize outliers using boxplots and remove them

# Set the plot size
plt.figure(figsize=(14, 6))

# Create boxplots for all numeric columns
sns.boxplot(data=df)

plt.title("Boxplot of all columns")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Function to remove outliers using IQR for all numeric columns
def remove_outliers_iqr(data):
    cleaned = df.copy()
    for col in cleaned.select_dtypes(include=['number']).columns:
        Q1 = cleaned[col].quantile(0.25)
        Q3 = cleaned[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        # Filter values within bounds
        cleaned = cleaned[(cleaned[col] >= lower) & (cleaned[col] <= upper)]
    return cleaned

df_clean = remove_outliers_iqr(df)
```

```python
print("Original shape:", df.shape)
print("Cleaned shape :", df_clean.shape)

# Optional: visualize again after cleaning
plt.figure(figsize=(14, 6))
sns.boxplot(data=df_clean)
plt.title("Boxplot after removing outliers")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
